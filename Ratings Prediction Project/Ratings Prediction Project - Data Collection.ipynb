```python
element_list = []
  
for page in range(1,65,1):
    
    page_url = "https://www.flipkart.com/hp-pavilion-ryzen-5-hexa-core-5600h-8-gb-512-gb-ssd-windows-10-4-graphics-nvidia-geforce-gtx-1650-144-hz-15-ec2004ax-gaming-laptop/product-reviews/itm98c94bbf9bc20?pid=COMG5GZXPWMGTNWS&lid=LSTCOMG5GZXPWMGTNWSQE9WVW&marketplace=FLIPKART&page=1"+ str(page)
    rating = driver.find_elements_by_xpath('//div[@class="_3LWZlK _1BLPMq"]')
    review_summary = driver.find_elements_by_xpath('//p[@class="_2-N8zT"]')
    full_review = driver.find_elements_by_xpath('//div[@class="t-ZTKy"]')
    for i in range(len(rating)):
        element_list.append([rating[i].text, review_summary[i].text, full_review[i].text])

R=640
r = [i[0] for i in element_list[ : R]]

RS=640
rs = [i[1] for i in element_list[ : RS]]

FR=640
fr = [i[2] for i in element_list[ : FR]]

lap=pd.DataFrame()
lap['Rating']=r
lap['Review_Summary']=rs
lap['Full_Review']=fr
```
