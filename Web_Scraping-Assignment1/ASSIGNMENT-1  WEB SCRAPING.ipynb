```python
# 1) Write a python program to display all the header tags from wikipedia.org.
# Answer:-
```

```python
# We need to send get request to webpage server to get the source code of the page.
Wiki = requests.get('https://en.wikipedia.org/wiki/Main_Page')
# Then download all the Page Content
Content = BeautifulSoup(Wiki.content)
# For Scraping all the Headers
# First create empty list & then append all the headers
Allheaders = []
for i in Content.find_all('span',class_="mw-headline"):
    Allheaders.append(i.text)
Allheaders
```

```python
# 2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.
# Answer:-
```

```python
# We need to send get request to webpage server to get the source code of the page.
imdb = requests.get('https://www.imdb.com/chart/top/?sort=ir,desc&mode=simple&page=1')
# Download all the Page Content
imdbcontent = BeautifulSoup(imdb.content)
# For Scraping name, rating, year of release
# First create empty list & then append
name = []
for i in imdbcontent.find_all('td',class_="titleColumn"):
    name.append(i.text.replace('\n',' '))

rating = []
for i in imdbcontent.find_all('td',class_="ratingColumn imdbRating"):
    rating.append(i.text.replace('\n',' '))

year = []
for i in imdbcontent.find_all('span',class_="secondaryInfo"):
    year.append(i.text)

df = pd.DataFrame({'Movies':name,'Ratings':rating,'Year_of_release':year})
pd.set_option('display.max_rows',None)

# IMDB’s Top rated 100 movies
df.iloc[:100]
```

```python
# 5) Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:
# a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.
# b) Top 10 ODI Batsmen along with the records of their team and rating.
# c) Top 10 ODI bowlers along with the records of their team and rating.
# Answers:-
```

```python
# a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

# banner text data
t = requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
c = BeautifulSoup(t.content)
a = []
for i in c.find_all('span',class_="u-hide-phablet"):
    a.append(i.text)
del a[1:25]
b = []
for i in c.find_all('td',class_="rankings-block__banner--matches"):
    b.append(i.text)
d = []
for i in c.find_all('td',class_="rankings-block__banner--points"):
    d.append(i.text)
e = []
for i in c.find_all('td',class_="rankings-block__banner--rating u-text-right"):
    e.append(i.text.replace('\n',''))

# rest of the text data
ac = []
for i in c.find_all('span',class_="u-hide-phablet"):
    ac.append(i.text)
del ac[20:25]
del ac[0]

Match = []
for i in c.find_all('td',class_="u-center-text"):
    Match.append(i.text)
x = 2
Match = [Match[i] for i in range(0, len(Match), x)]

Point = []
for i in c.find_all('td',class_="table-body__cell u-center-text"):
    Point.append(i.text)
x = 2
Point = [Point[i] for i in range(1, len(Point), x)]

ef = []
for i in c.find_all('td',class_="table-body__cell u-text-right rating"):
    ef.append(i.text)
# Create DataFrame & Merge
import pandas as pd
Cricketmatch = pd.DataFrame({'Teams':a,'Matches':b,'Points':d,'Ratings':e})
Cricketmatch1 = pd.DataFrame({'Teams':ac,'Matches':Match,'Points':Point,'Ratings':ef})
pd.concat([Cricketmatch,Cricketmatch1])
index = pd.Index(range(0, 20, 1))
dfcricket = pd.concat([Cricketmatch,Cricketmatch1]).set_index(index)
dfcricket.iloc[:10]
```

```python
# b) Top 10 ODI Batsmen along with the records of their team and rating.

we = requests.get('https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting')
ce = BeautifulSoup(we.content)
# Banner data
ae = []
for i in ce.find_all('div',class_="rankings-block__banner--name-large"):
    ae.append(i.text)
be = []
for i in ce.find_all('div',class_="rankings-block__banner--nationality"):
    be.append(i.text.replace('\n',''))
de = []
for i in ce.find_all('div',class_="rankings-block__banner--rating"):
    de.append(i.text)

# rest of the data
ac = []
for i in ce.find_all('td',class_="table-body__cell rankings-table__name name"):
    ac.append(i.text.replace('\n',''))
bc = []
for i in ce.find_all('span',class_="table-body__logo-text"):
    bc.append(i.text)
dc = []
for i in ce.find_all('td',class_="table-body__cell rating"):
    dc.append(i.text)

# Create DataFrame & Merge
batsman = pd.DataFrame({'Players':ae,'Teams':be,'Ratings':de})
batsman1 = pd.DataFrame({'Players':ac,'Teams':bc,'Ratings':dc})
pd.concat([batsman,batsman1])
index = pd.Index(range(0, 100, 1))
dfplayer = pd.concat([batsman,batsman1]).set_index(index)
dfplayer.iloc[:10]
```

```python
# c) Top 10 ODI bowlers along with the records of their team and rating.

wd = requests.get('https://www.icc-cricket.com/rankings/mens/player-rankings/odi/bowling')
cd = BeautifulSoup(wd.content)
# Banner data
ad = []
for i in cd.find_all('div',class_="rankings-block__banner--name-large"):
    ad.append(i.text)
bd = []
for i in cd.find_all('div',class_="rankings-block__banner--nationality"):
    bd.append(i.text.replace('\n',''))
dd = []
for i in cd.find_all('td',class_="u-text-left"):
    dd.append(i.text.replace('\n',''))

# Rest data
wd = requests.get('https://www.icc-cricket.com/rankings/mens/player-rankings/odi/bowling')
cd = BeautifulSoup(wd.content)
af = []
for i in cd.find_all('td',class_="table-body__cell rankings-table__name name"):
    af.append(i.text.replace('\n',''))
bf = []
for i in cd.find_all('span',class_="table-body__logo-text"):
    bf.append(i.text)
df = []
for i in cd.find_all('td',class_="table-body__cell rating"):
    df.append(i.text)

# Create DataFrame & Merge
bowler = pd.DataFrame({'Players':ad,'Teams':bd,'Ratings':dd})
bowler1 = pd.DataFrame({'Players':af,'Teams':bf,'Ratings':df})
pd.concat([batsman,batsman1])
index = pd.Index(range(0, 100, 1))
dfbowler = pd.concat([bowler,bowler1]).set_index(index)
dfbowler.iloc[:10]
```

```python
# b) Top 10 women’s ODI Batting players along with the records of their team and rating.

wd = requests.get('https://www.icc-cricket.com/rankings/womens/player-rankings/odi/batting')
cv = BeautifulSoup(wd.content)
# Banner data
al = []
for i in cv.find_all('div',class_="rankings-block__banner--name-large"):
    al.append(i.text.replace('\n',''))
bl = []
for i in cv.find_all('div',class_="rankings-block__banner--nationality"):
    bl.append(i.text.replace('\n',''))
dl = []
for i in cv.find_all('div',class_="rankings-block__banner--rating"):
    dl.append(i.text)

# Rest data
av = []
for i in cv.find_all('td',class_="table-body__cell rankings-table__name name"):
    av.append(i.text.replace('\n',''))
bv = []
for i in cv.find_all('td',class_="table-body__cell nationality-logo rankings-table__team"):
    bv.append(i.text.replace('\n',''))
dv = []
for i in cv.find_all('td',class_="table-body__cell rating"):
    dv.append(i.text)

# Create DataFrame & Merge
batswomen = pd.DataFrame({'Players':al,'Teams':bl,'Ratings':dl})
batswomen1 = pd.DataFrame({'Players':av,'Teams':bv,'Ratings':dv})
pd.concat([batswomen,batswomen1])
index = pd.Index(range(0, 100, 1))
dfbatswomen = pd.concat([batswomen,batswomen1]).set_index(index)
dfbatswomen.iloc[:10]
```

```python
# 7) Write a python program to scrape details of all the posts from coreyms.com. 
# Scrape the heading, date, content and the code for the video from the link for the youtube video from the post.
# Answer:-
```

```python
coreyms = requests.get('https://coreyms.com/')
content = BeautifulSoup(coreyms.content)

# Scrape the heading, date, content and the code for the video
Heading = []
for i in content.find_all('header',class_="entry-header"):
    Heading.append(i.text.replace('\n',' '))

Date = []
for i in content.find_all('time',class_="entry-time"):
    Date.append(i.text)
    
Cont = []
for i in content.find_all('div',class_="entry-content"):
    Cont.append(i.text.replace('\n',' '))
    
Video = []
for i in content.find_all('span',class_="embed-youtube"):
    Video.append(i)
Video.insert(4,"Video URL Not Available")
    
dfcoreyms = pd.DataFrame({'Name':Heading,'Date_Of_Post':Date,'Content':Cont,'VideoURL':Video})
dfcoreyms
```

```python
# 9) Write a python program to scrape mentioned details from dineout.co.in : 
# i) Restaurant name ii) Cuisine iii) Location iv) Ratings v) Image URL
# Answer:-
```

```python
Restaurant = requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
scrapecontent = BeautifulSoup(Restaurant.content)

restaurantname = []
for i in scrapecontent.find_all('a',class_="restnt-name ellipsis"):
    restaurantname.append(i.text)
    
restaurantcuisine = []
for i in scrapecontent.find_all('span',class_="double-line-ellipsis"):
    restaurantcuisine.append(i.text)
    
restaurantloc = []
for i in scrapecontent.find_all('div',class_="restnt-loc ellipsis"):
    restaurantloc.append(i.text)
    
restaurantrating = []
for i in scrapecontent.find_all('div',class_="restnt-rating rating-4"):
    restaurantrating.append(i.text)
    
restaurantimages = []
for i in scrapecontent.find_all("img",class_="no-img"):
    restaurantimages.append(i['data-src'])
    
Dineout = pd.DataFrame({'Restaurants':restaurantname,'Cuisine':restaurantcuisine,'Location':restaurantloc,'Ratings':restaurantrating,'Image_url':restaurantimages})
Dineout
```
