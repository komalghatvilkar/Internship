```python
element_list = []
  
for page in range(0, 3, 1):
    page_url = "https://www.amazon.in/s?k=iphone&crid=28XXOVK888B50&sprefix=iphone%2Caps%2C485&ref=nb_sb_noss_1"+ str(page)
    brand = driver.find_elements_by_xpath('//span[@class="a-size-large product-title-word-break"]')
    name= driver.find_elements_by_xpath('//h2[@class="a-size-mini a-spacing-none a-color-base s-line-clamp-2"]')
    price = driver.find_elements_by_xpath('//span[@class="a-price"]')
    exchange = driver.find_elements_by_xpath('//a[@class="a-size-small a-link-normal a-text-normal"]')
    delivery = driver.find_elements_by_xpath('//div[@class="a-section a-spacing-none a-spacing-top-micro"]')
    availability = driver.find_elements_by_xpath('//span[@class="a-size-medium a-color-success"]')
    url = driver.find_elements_by_xpath('//a[@class="a-link-normal s-no-outline"]')
    for i in range(len(brand)):
        element_list.append([brand[i].text,name[i].text,price[i].text,exchange[i].text,delivery[i].text,availability[i].text,url[i].text])

br=200
b = [i[0] for i in element_list[ :br]]

na=200
n = [i[1] for i in element_list[ :na]]

pr=200
p = [i[2] for i in element_list[ :pr]]

ex=200
e = [i[3] for i in element_list[ :ex]]

di=200
d = [i[4] for i in element_list[ :di]]

av=200
a = [i[5] for i in element_list[ :av]]

ur=200
u = [i[6] for i in element_list[ :ur]]

df=pd.DataFrame()
df['Brand Name']=b
df['Name Of Product']=n
df['Price']=p
df['Return/Exchange']=e
df['Expected Delivery']=d
df['Availability']=a
df['Product URL']=u
df
```
